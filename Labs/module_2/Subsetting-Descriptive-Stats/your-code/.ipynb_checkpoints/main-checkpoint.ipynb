####  What is the average purchase price?

```python
# your code here => 3.12€

orders['UnitPrice'].mean()
```

#### What are the highest and lowest purchase prices? 

```python
# your code here
orders['UnitPrice'].max()
orders['UnitPrice'].min()
```

#### Select all the customers from Spain.
**Hint**: Remember that you are not asked to find orders from Spain but customers. A customer might have more than one order associated. 

```python
# your code here
orders[orders['Country']=="Spain"]
```

#### How many customers do we have in Spain?

```python
# your code here
orders[orders['Country']=="Spain"]['CustomerID'].nunique()
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here
filter = orders[['CustomerID','Quantity']].groupby(['CustomerID']).sum()

filter[filter['Quantity']>50]
```

#### Select orders from Spain that include more than 50 items.

```python
# your code here
orders[(orders['Country']=="Spain") & (orders['Quantity']>50)]
```

#### Select all free orders.

```python
# your code here
orders[orders['amount_spent']==0]
```

#### Select all orders whose description starts with `lunch bag`.
**Hint**: use string functions.

```python
# your code here
orders[orders['Description'].str.contains("lunch bag*")]
```

#### Select all `lunch bag` orders made in 2011.

```python
# your code here
orders[(orders['Description'].str.contains("lunch bag*")) & (orders['year']==2011)]
```

#### Show the frequency distribution of the amount spent in Spain.

```python
# your code here

#orders[orders['Country']=="Spain"]['amount_spent'].value_counts()

orders[orders['Country']=="Spain"]['amount_spent'].plot()
```

#### Select all orders made in the month of August.

```python
# your code here
orders[orders['month']== 8]
```

#### Find the number of orders made by each country in the month of August.
**Hint**: Use value_counts().

```python
# your code here
orders[orders['month']== 8]['Country'].value_counts()
```

#### What's the  average amount of money spent by country?

```python
# your code here
orders[['Country','amount_spent']].groupby(['Country']).mean()
```

#### What's the most expensive item?

```python
# your code here
orders[orders['UnitPrice'] == orders['UnitPrice'].max()]
```

```python
orders['UnitPrice'].max()
```

#### What is the average amount spent per year?

```python
# your code here
orders[['year','amount_spent']].groupby(['year']).mean()
```
